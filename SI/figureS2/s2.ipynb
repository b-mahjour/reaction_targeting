```python
import pandas as pd

dataC1 = pd.read_excel("../../output_datasets/aryl-aryl-cc.xlsx")
dataC2 = pd.read_excel("../../output_datasets/aryl-alkyl-Cc.xlsx")
dataC3 = pd.read_excel("../../output_datasets/alkyl-alkyl-CC.xlsx")

dataN1 = pd.read_excel("../../output_datasets/alkyl-alkyl-CN.xlsx")
dataN2 = pd.read_excel("../../output_datasets/alkyl-aryl-Cn.xlsx")
dataN3 = pd.read_excel("../../output_datasets/aryl-alkyl-cN.xlsx")
dataN4 = pd.read_excel("../../output_datasets/aryl-aryl-cn.xlsx")

dataO1 = pd.read_excel("../../output_datasets/alkyl-alkyl-CO.xlsx")
dataO2 = pd.read_excel("../../output_datasets/aryl-alkyl-cO.xlsx")
```

```python
synths = {}
```

```python
arrs = [dataC1, dataC2, dataC3, dataN1, dataN2, dataN3, dataN4, dataO1, dataO2]
```

```python
for k in arrs:
    l = k.groupby("synthon_a_building_block_smiles")
    for i,j in l:
        if i not in synths:
            synths[i] = 0
        synths[i] = synths[i] + len(j)
    l = k.groupby("synthon_b_building_block_smiles")
    for i,j in l:
        if i not in synths:
            synths[i] = 0
        synths[i] = synths[i] + len(j)
```

```python
hits = []
for k in synths:
    if synths[k] == 1:
        continue
    hits.append((synths[k],k))
hits.sort(reverse=True)
```

```python
all_hits = []

for k in hits:
    if len(all_hits) > 200:
        continue
    all_hits.append(k[0])
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(3.7,1))
ax.bar(range(0,len(all_hits)), np.log10(all_hits), color='#cac9c9')
ax.bar(0, np.log10(3182), color='#757BBB', width=1, zorder=2)
ax.bar(7, np.log10(1360), color='#757BBB', width=1, zorder=2)
ax.bar(20, np.log10(673), color='#757BBB', width=1, zorder=2)
ax.bar(39, np.log10(406), color='#757BBB', width=1, zorder=2)
ax.bar(147, np.log10(132), color='#757BBB', width=1, zorder=2)
ax.bar(192, np.log10(102), color='#757BBB', width=1, zorder=2)

ax.set_xticks([50, 100, 150])
ax.set_xticklabels(ax.get_xticks(), fontsize=6, fontfamily='arial')
ax.set_yticks([0, 1, 2, 3, 4])
ax.set_ylim([0, 4])
ax.set_yticklabels(ax.get_yticks(), fontsize=6, fontfamily='arial')

# ax.set_yticks([0, 2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000, 30000])
# ax.set_yticklabels([0, 2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000, 30000], fontsize=6, fontfamily='arial')
# ax.set_ylim([0,8000])

ax.set_ylabel("log(frequency)", fontsize=6, fontfamily='arial')
ax.set_xlabel("building block", fontsize=6, fontfamily='arial')
ax.set_title("building block occurrence in\n single step reactions to DrugBank compounds", fontsize=6, fontfamily='arial')

plt.subplots_adjust(hspace=0, wspace=0)
plt.margins(0,0)

plt.savefig(f"component_a.png", dpi=900, bbox_inches='tight', pad_inches = 0)
```
